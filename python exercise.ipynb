```python
def is_even(num):
    return num %2 == 0

def count_even(numbers):
    count = 0

    for num in numbers:
        if is_even(num):
            count+= 1

    return count
```

```python
count_even([1,2,3,4,5,6,8])
```

```python
def animal_cracker(text):
    words = text.split()
    return words[0][0].lower() == words[1][0].lower()
```

```python
def master_yoda(text):
    result = text.split()
    reversed_words = result[::-1]
    join = " ".join(reversed_words)
    return join
```

```python
master_yoda('I am home')
```

```python
def almost_there(n):

    distance_from_100 = abs(n - 100)
    distance_from_200 = abs(n - 200)

    if distance_from_100 <= 10 or distance_from_200 <= 10:
        return True
    else:
        return False
```

```python
almost_there(105)
```

```python
almost_there(204)
```

```python
def fizz_buzz(n):
    for num in range(1, n + 1):
        if num % 3 == 0 and num % 5 == 0:
            print("FizzBuzz")
        elif num % 3 == 0:
            print("Fizz")
        elif num % 5 == 0:
            print("Buzz")
        else:
            print(num)
```

```python
fizz_buzz(15)
```
